<h3>
14. Parameter passing mechanisms.
</h3>
<table>
            <tr>
                <th>Concept</th>
                <th>Definition</th>
            </tr>
            <tr>
                <th>Pass By Value</th>
                <td>Pass a copy of the value to the subroutine. Changes made to any formal parameters inside the function do not modify the passed values, as they are a copy. Java is only call by value.</td>
            </tr>
            <tr>
                <th>Pass By Reference</th>
                <td>References the original var, so changes made to formal parameters are reflected t o the actual data, or actual formal parameter passed.</td>
            </tr>
            <tr>
                <th>Pass By Result</th>
                <td>Value of formal parameter gets transmitted back to the actual parameter.</td>
            </tr>
            <tr>
                <th>Pass By Value-Result</th>
                <td>Pass by value + pass by result, formal parameter gets transmitted back to the actual parmeter.</td>
            </tr>
    </table>
